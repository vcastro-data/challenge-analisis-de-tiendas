```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Leer CSV desde las URLs
tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Asignar nombre de tienda
tienda1["Tienda"] = "tienda_1"
tienda2["Tienda"] = "tienda_2"
tienda3["Tienda"] = "tienda_3"
tienda4["Tienda"] = "tienda_4"

# Unir todo en un solo dataframe
data = pd.concat([tienda1, tienda2, tienda3, tienda4], ignore_index=True)

# Mostrar primeros datos
data.head()
```

#1. Análisis de facturación



```python
ingreso_por_tienda = data.groupby("Tienda")["Precio"].sum()
print("Ingreso total por tienda:")
print(ingreso_por_tienda)
```

```python
import matplotlib.pyplot as plt

ingreso_por_tienda = data.groupby("Tienda")["Precio"].sum()

plt.figure()
ingreso_por_tienda.plot(kind="bar", title="Ingresos Totales por Tienda")
plt.xlabel("Tienda")
plt.ylabel("Ingresos Totales")
plt.tight_layout()
plt.show()
```

podemos observar que la tienda con mayor ingreso es la numero 1 y la tienda con menor ingreso es la  numero 4

# 2. Ventas por categoría

```python
ventas_por_categoria = data["Categoría del Producto"].value_counts()
print("Ventas por categoría:")
print(ventas_por_categoria)
```

```python
ventas_por_categoria = data["Categoría del Producto"].value_counts()

plt.figure()
ventas_por_categoria.plot(kind="bar", title="Ventas por Categoría de Producto")
plt.xlabel("Categorías")
plt.ylabel("Cantidad Vendida")
plt.tight_layout()
plt.show()
```

La categoria con mayor numero de ventas es la de muebles sobrepasando las 1.750 unidades, en los articulos con menor numeros de articulos vendidos son los instrumentos musicales, los libros y los articulos para el hogar estando casi en el mismo rango de 750 unidades

# 3. Calificación promedio de la tienda


```python
valoracion_media = data.groupby("Tienda")["Calificación"].mean()
print("Valoración media por tienda:")
print(valoracion_media)
```

```python
valoracion_media = data.groupby("Tienda")["Calificación"].mean()

plt.figure()
valoracion_media.plot(kind="bar", title="Valoración Media por Tienda")
plt.xlabel("Tienda")
plt.ylabel("Calificación Promedio")
plt.ylim(0,5)  # rango de calificación
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
productos_contados = data["Producto"].value_counts()

mas_vendido = productos_contados.idxmax()
menos_vendido = productos_contados.idxmin()

print("Producto más vendido:", mas_vendido)
print("Producto menos vendido:", menos_vendido)
```

```python
top_productos = data["Producto"].value_counts().head(10)

plt.figure()
top_productos.plot(kind="bar", title="Top 10 Productos Más Vendidos")
plt.xlabel("Producto")
plt.ylabel("Cantidad Vendida")
plt.tight_layout()
plt.show()
```
